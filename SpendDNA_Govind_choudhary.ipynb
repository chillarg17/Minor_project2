```python
import pandas as pd
import numpy as np
df=pd.read_csv('Data set for DADS June.csv')
print("dataset imported successfully")
df
print("-" * 60)

print("Shape of Dataset :", df.shape)

print("\nFirst Five Rows")
print(df.head())

df_clean = df.copy()

# Dataset Information
print("=" * 60)
print("DATASET INFORMATION")
print("=" * 60)

print(df.info())

print("\n")

print("=" * 60)
print("COLUMN NAMES")
print("=" * 60)

print(df.columns)

print("\n")

print("=" * 60)
print("DATA TYPES")
print("=" * 60)

print(df.dtypes)
```

```python
#converting dates(4 formats)

df_clean['Date'] = pd.to_datetime(
    df_clean['Date'],
    errors='coerce',
    dayfirst = True
)
print(df_clean)
```

```python
#converting amount(3 formats)
df_clean["Amount"] = (
    df_clean["Amount"]
    .astype(str)
    .str.replace("₹", "", regex=False)
    .str.replace("Rs.", "", regex=False)
    .str.replace("-", "", regex = False)
     .str.replace("Rs", "", regex=False)
    .str.replace(",", "", regex=False)
    .str.strip()
)



df_clean["Amount"] = pd.to_numeric(
    df_clean["Amount"],
    errors="coerce"
)
print(df_clean['Amount'].dtypes)

df_clean.head(200)
```

```python
# standerdize transection type

df_clean["Type"] = (
    df_clean["Type"]
    .astype(str)
    .str.strip()
    .str.lower()
)

df_clean["Type"] = df_clean["Type"].replace({
    "dr": "debit",
    "Debit": "debit",
    "Credit": "credit",
    "cr": "credit"
})

print(df_clean["Type"].value_counts())
```

```python
# duplicates removed
duplicates_removed = df_clean.duplicated().sum()

df_clean = df_clean.drop_duplicates()

print("Duplicates Removed :", duplicates_removed)


invalid_dates = df_clean["Date"].isna().sum()
invalid_amounts = df_clean["Amount"].isna().sum()

df_clean = df_clean.dropna(
    subset=["Date", "Amount"]
)

print("Invalid Dates :", invalid_dates)
print("Invalid Amounts :", invalid_amounts)

print("=" * 60)
print("CLEAN DATASET INFO")
print("=" * 60)

print(df_clean.info())

print("\n")

print(df_clean.head())

print("\n")

print("Final Shape :", df_clean.shape)
```

```python
# Vendor Extraction Function:    AI-Assisted

def extract_vendor(description):

    description = str(description).upper()

    # ATM Withdrawal
    if "ATM" in description:
        return "Cash Withdrawal"

    # Friend-to-Friend Transfer
    if "OKAXIS" in description or "OKHDFC" in description or "OKSBI" in description:

        if (
            "SWIGGY" not in description and
            "ZOMATO" not in description and
            "AMAZON" not in description and
            "PAYTM" not in description
        ):
            return "P2P Transfer"

    # Vendor Matching
    for vendor, keywords in vendor_dictionary.items():

        for keyword in keywords:

            if keyword in description:
                return vendor

    return "Uncategorised"
```

```python
# Create Vendor Column

df_clean["vendor_clean"] = df_clean["Description"].apply(extract_vendor)

print("Vendor Extraction Completed Successfully!")
```

```python
# FEATURE 3 : CATEGORY TAGGER

category_dictionary = {

    "Swiggy": "Food Delivery",
    "Zomato": "Food Delivery",

    "Zepto": "Quick Commerce",
    "Blinkit": "Quick Commerce",
    "Instamart": "Quick Commerce",

    "Amazon": "E-Commerce",
    "Flipkart": "E-Commerce",
    "Myntra": "E-Commerce",
    "Nykaa": "E-Commerce",

    "Uber": "Transport",
    "Ola": "Transport",
    "Rapido": "Transport",

    "Netflix": "Entertainment",
    "Spotify": "Entertainment",
    "Hotstar": "Entertainment",
    "BookMyShow": "Entertainment",

    "Jio": "Recharge",
    "Airtel": "Recharge",
    "Vodafone": "Recharge",

    "Electricity": "Bills",
    "Water Bill": "Bills",

    "Groww": "Investment",
    "Zerodha": "Investment",

    "Paytm": "Wallet",

    "Cash Withdrawal": "Cash Withdrawal",

    "P2P Transfer": "Personal Transfer",

    "Uncategorised": "Others"
}

df_clean["Category"] = df_clean["vendor_clean"].map(category_dictionary)

df_clean["Category"] = df_clean["Category"].fillna("Others")

print("=" * 60)
print("CATEGORY DISTRIBUTION")
print("=" * 60)

print(df_clean["Category"].value_counts())
```

```python
#SPENDING OVERVIEW
total_transactions = len(df_clean)
total_credit = df_clean[df_clean["Type"] == "credit"]["Amount"].sum()

total_debit = df_clean[df_clean["Type"] == "debit"]["Amount"].sum()

net_savings = total_credit - total_debit
if total_credit > 0:
  savings_rate = (net_savings / total_credit) * 100
else:
  saving_rate = 0
category_spending = (
    df_clean[df_clean["Type"] == "debit"]
    .groupby("Category")["Amount"]
    .sum()
    .sort_values(ascending=False)
)

top_vendors = (
    df_clean[df_clean["Type"] == "debit"]
    .groupby("vendor_clean")["Amount"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

print("=" * 70)
print("SPENDING OVERVIEW")
print("=" * 70)

print(f"Total Transactions : {total_transactions}")
print(f"Total Credit       : ₹{total_credit:,.2f}")
print(f"Total Debit        : ₹{total_debit:,.2f}")
print(f"Net Savings        : ₹{net_savings:,.2f}")
print(f"Savings Rate       : {savings_rate:.2f}%")

print("\n")

print("=" * 70)
print("CATEGORY WISE SPENDING")
print("=" * 70)

print(category_spending)

print("\n")

print("=" * 70)
print("TOP 10 VENDORS")
print("=" * 70)

print(top_vendors)
```

```python
#MONTHLY TREND ANALYSIS
# Create Month Column
df_clean["Month"] = df_clean["Date"].dt.strftime("%B")

# Month Order
month_order = [
    "January", "February", "March", "April",
    "May", "June", "July", "August",
    "September", "October", "November", "December"
]

# Convert into Ordered Category
df_clean["Month"] = pd.Categorical(
    df_clean["Month"],
    categories=month_order,
    ordered=True
)

# Monthly Spending Trend
monthly_trend = (
    df_clean[df_clean["Type"] == "debit"]
    .pivot_table(
        values="Amount",
        index="Month",
        columns="Category",
        aggfunc="sum",
        fill_value=0
    )
)

monthly_trend = monthly_trend.sort_index()

print("=" * 70)
print("MONTHLY SPENDING TREND")
print("=" * 70)

monthly_trend.round(2)
```

```python
#TIME OF DAY ANALYSIS
# Extract Hour
df_clean["Hour"] = df_clean["Time"].str[:2].astype(int)

# Time Slots
def time_period(hour):

    if 5 <= hour < 12:
        return "Morning"

    elif 12 <= hour < 17:
        return "Afternoon"

    elif 17 <= hour < 21:
        return "Evening"

    else:
        return "Night"

# Create Time Period Column
df_clean["Time Period"] = df_clean["Hour"].apply(time_period)

# Time Period Distribution
time_period_spending = (
    df_clean[df_clean["Type"] == "debit"]
    .groupby("Time Period")["Amount"]
    .sum()
    .sort_values(ascending=False)
)

print("=" * 70)
print("TIME OF DAY SPENDING")
print("=" * 70)

print(time_period_spending)

# Food Delivery by Time
food_delivery = df_clean[
    (df_clean["Category"] == "Food Delivery") &
    (df_clean["Type"] == "debit")
]

food_time = (
    food_delivery
    .groupby("Time Period")
    .size()
)

print("\n")

print("=" * 70)
print("FOOD DELIVERY ORDERS BY TIME")
print("=" * 70)

print(food_time)

# Late Night Food Orders
late_night_orders = food_delivery[
    food_delivery["Hour"] >= 21
]

total_food_orders = len(food_delivery)

late_night_count = len(late_night_orders)

if total_food_orders > 0:
    late_night_percentage = (
        late_night_count / total_food_orders
    ) * 100
else:
    late_night_percentage = 0

print("\n")

print("=" * 70)
print("LATE NIGHT FOOD ANALYSIS")
print("=" * 70)

print(f"Total Food Orders      : {total_food_orders}")
print(f"Late Night Orders      : {late_night_count}")
print(f"Late Night Percentage  : {late_night_percentage:.2f}%")
```

```python
#ANOMALY DETECTION                                           AI Assisted
# Consider only debit transactions
debit_df = df_clean[df_clean["Type"] == "debit"].copy()

# Mean and Standard Deviation for each Category
category_mean = debit_df.groupby("Category")["Amount"].transform("mean")
category_std = debit_df.groupby("Category")["Amount"].transform("std")

# Replace 0 standard deviation with NaN to avoid division by zero
category_std = category_std.replace(0, np.nan)

# Manual Z-Score
debit_df["Z-Score"] = (debit_df["Amount"] - category_mean) / category_std

# Fill NaN with 0
debit_df["Z-Score"] = debit_df["Z-Score"].fillna(0)

# Detect Anomalies
anomalies = debit_df[debit_df["Z-Score"] > 2]

print("=" * 70)
print("ANOMALY DETECTION")
print("=" * 70)

print(f"Total Debit Transactions : {len(debit_df)}")
print(f"Anomalies Found          : {len(anomalies)}")

print("\nTop Anomalies\n")

print(
    anomalies[
        ["Date", "vendor_clean", "Category", "Amount", "Z-Score"]
    ].sort_values("Z-Score", ascending=False)
)

#SPENDING ARCHETYPE DETECTION                                        AI assisted
archetypes = []

# Total Debit
total_debit = df_clean[df_clean["Type"] == "debit"]["Amount"].sum()

# Category Spending
category_spending = (
    df_clean[df_clean["Type"] == "debit"]
    .groupby("Category")["Amount"]
    .sum()
)

# Foodie
food_spend = category_spending.get("Food Delivery", 0)

if total_debit > 0 and (food_spend / total_debit) >= 0.20:
    archetypes.append("🍔 Foodie")

# Investor
investment_spend = category_spending.get("Investment", 0)

if total_debit > 0 and (investment_spend / total_debit) >= 0.15:
    archetypes.append("📈 Investor")

# Shopaholic
shopping = category_spending.get("E-Commerce", 0)

if total_debit > 0 and (shopping / total_debit) >= 0.20:
    archetypes.append("🛍️ Shopaholic")

# Frequent Traveler
transport = category_spending.get("Transport", 0)

if total_debit > 0 and (transport / total_debit) >= 0.10:
    archetypes.append("🚖 Frequent Traveler")

# Entertainment Lover
entertainment = category_spending.get("Entertainment", 0)

if total_debit > 0 and (entertainment / total_debit) >= 0.10:
    archetypes.append("🎬 Entertainment Lover")

# Late Night Snacker
food_orders = df_clean[
    (df_clean["Category"] == "Food Delivery") &
    (df_clean["Type"] == "debit")
]

late_night_orders = food_orders[food_orders["Hour"] >= 21]

if len(food_orders) > 0:
    late_night_percentage = (len(late_night_orders) / len(food_orders)) * 100
else:
    late_night_percentage = 0

if late_night_percentage >= 30:
    archetypes.append("🌙 Late Night Snacker")

# Cash Lover
cash = category_spending.get("Cash Withdrawal", 0)

if total_debit > 0 and (cash / total_debit) >= 0.15:
    archetypes.append("💵 Cash Lover")

# Digital Wallet User
wallet = category_spending.get("Wallet", 0)

if total_debit > 0 and (wallet / total_debit) >= 0.10:
    archetypes.append("📱 Digital Wallet User")

# YOLO Spender
if len(anomalies) >= 5:
    archetypes.append("💸 YOLO Spender")

# Saver
if total_credit > 0 and savings_rate >= 30:
    archetypes.append("💰 Saver")

# Default
if len(archetypes) == 0:
    archetypes.append("🙂 Balanced Spender")

print("=" * 70)
print("SPENDING ARCHETYPES")
print("=" * 70)

for archetype in archetypes:
    print(archetype)

print("\nTotal Archetypes :", len(archetypes))
```
